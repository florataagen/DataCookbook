##### Let's create a line plot using seaborn

Below is a line plot with date on the x-axis and monthly mean total sunspot number on the y-axis. 

- the set_theme( ) method sets the theme of the plot to one of seaborn's defined themes.
- the lineplot( ) method creates a lineplot according to the supplied x-axis, y-axis, and data source.
- the tick_params( ) method rotates the x-axis and y-axis labels by the specified degrees.
- the title method adds a title to the plot.
- the show( ) method displays the plot.


```python
sns.set_theme(style="darkgrid")

ax = sns.lineplot(x="Date", y="Monthly Mean Total Sunspot Number", data=df)
ax.tick_params(rotation = 60)

plt.title("Monthly Mean Total Sunspot Number from 1749 - 2021")
plt.show()
```
